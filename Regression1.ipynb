```python
# Assumption
## 1. That there exists a relationship between Olympic year and winning time.
## 2. That this relationship is linear (i.e. a straight line).
## 3. That this relationship will continue into the future.

# fit a straight line
w_0 = 36.4164559025
w_1 = -0.013330885711

# generate new x to plot the fitted line.
# Note better not to use the original x!
x_test = np.linspace(1880, 2020, 100) 
f_test = w_0 + w_1 * x_test

#plot the fitted data
plt.plot(x_test, f_test, 'b-', linewidth=2)
plt.plot(2012, w_0 + w_1 * 2012, 'ro')

#draw a scatter plot
plt.scatter(x,t)
plt.xlabel('years')
plt.ylabel('Time (seconds)')
```
